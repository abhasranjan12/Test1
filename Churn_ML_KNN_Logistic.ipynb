```python
df3[['MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies']]=df1[['MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies']]
```

```python
df3.drop(['gender'],axis=1)
```

```python
df3=df3.merge(df2)
```

```python
df3.shape
```

```python
df3.columns
```

```python
X=df3[ ['tenure', 'PhoneService', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'gender',
       'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
       'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
       'SeniorCitizen', 'Partner', 'Dependents']]
y=df3[['Churn']]
```

```python
X=pd.get_dummies(X)
y.dtypes
```

```python
X.shape
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df3[['Churn']]=le.fit_transform(df3[['Churn']])
```

```python
df3
y.dtypes
```

```python
y=df3[['Churn']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
le=LogisticRegression()
le.fit(X_train,y_train)
```

```python
from sklearn import metrics
```

```python
y_pred=le.predict(X_test)
y_pred_train=le.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
print(confusion_matrix(y_train,y_pred_train))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(classification_report(y_train,y_pred_train))
```
