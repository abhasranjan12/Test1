```python
df1['Total classes'].fillna(value=0,inplace=True)
df1['Crossed red mark'].fillna(value=0.0,inplace=True)
df1['Dropout days'].fillna(value=0.0,inplace=True)
```

```python
df1.isnull().sum()
```

# After Removing null values

```python
df1
```

```python
df1['Start Timing']=df1['Timing'].str.split('-',expand=True)[0].str.slice(1,10)
df1['End Timing']=df1['Timing'].str.split('-',expand=True)[1].str.slice(0,7)
```

```python
df1.drop(['Timing'],1,inplace=True)
```

```python
df1
```

```python
df1['Start_Day']=df1['Start Date'].str.split('-',expand=True)[0]
df1['Start_Month']=df1['Start Date'].str.split('-',expand=True)[1]
df1['Start_Year']=df1['Start Date'].str.split('-',expand=True)[2]
```

```python
df1.drop('Start Date',1,inplace=True)
```

```python
df1
```

```python
look_up = {'01': 'Jan', '02': 'Feb', '03': 'Mar', '04': 'Apr', '05': 'May',
            '06': 'Jun', '07': 'Jul', '08': 'Aug', '09': 'Sep', '10': 'Oct', '11': 'Nov', '12': 'Dec'}

df1['Start_Month'] = df1['Start_Month'].apply(lambda x: look_up[x])
df1
```

```python
df1['Actual_start_date']=df1['Device Count'].str.split('to',expand=True)[0]
df1['Actual_end_date']=df1['Device Count'].str.split('to',expand=True)[1]
df1['No_of_actual_classs']=df1['Actual_start_date'].str.split('2018',expand=True)[0]
```

```python
df1
```

```python
df1['Actual_start_date']=df1['Actual_start_date'].str.slice(2,16)
df1.drop(['Device Count'],1,inplace=True)
```
