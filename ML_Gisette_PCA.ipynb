```python
x_train=df.drop('target',axis=1,inplace=True)
```

```python
df=df.drop('Unnamed: 5000',axis=1)
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_scaled=sc.fit_transform(df)
x_train_scaled
```

```python
x_train_scaled.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=9)
```

```python
x_pca=pca.fit_transform(x_train_scaled)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pc=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9']
```

```python
plt.bar(pc,pca.explained_variance_ratio_)
```

```python
x_pca_df=pd.DataFrame(x_pca,columns=pc)
x_pca_df
```

```python
df.columns
```

```python
x_pca_df['Target']=df['0']
x_pca_df
```

```python
y=x_pca_df['Target']
y
```

```python
x_pca_df=x_pca_df.drop(['Target'],axis=1)
```

```python
x_pca_df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_pca_df,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
lgr=LogisticRegression()
lgr.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
y_pred=lgr.predict(x_test)
```

# Test Data Accuracy

```python
results=confusion_matrix(y_test,y_pred.round())
print(results)
accuracy_score(y_test,y_pred.round())
```

# Train Data Accuracy

```python
y_pred_train=lgr.predict(x_train)
```

```python
results=confusion_matrix(y_train,y_pred_train.round())
print(results)
accuracy_score(y_train,y_pred_train.round())
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
from sklearn import metrics
y_pred = knn.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred))
```
